```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.stats.diagnostic as diagnostic 
import matplotlib.pyplot as plt

plt.style.use('seaborn')
```

```python
dataset = pd.read_csv("dataset/adjusted.csv")
dataset.shape
```

```python
plt.hist(dataset["screen_time"], bins=20, facecolor="blue", alpha=0.7)
plt.xlabel("Screen Time")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(dataset["sleep_time"], bins=20, facecolor="blue", alpha=0.7)
plt.xlabel("Sleep Time")
plt.ylabel("Count")
plt.show()
```

```python
encoded_dataset = pd.read_csv("dataset/encoded.csv")
encoded_dataset.shape
```

```python
## create scatter plot of sleep time and screen time.

plt.scatter(x=encoded_dataset["screen_time"], y=encoded_dataset["sleep_time"])
plt.xlabel("Screen Time")
plt.ylabel("Sleep Time")
plt.show()
```
