```python
import pandas as pd
```

```python
#1 Encode datasets

raw = pd.read_csv("dataset/adjusted.csv")
raw.shape
```

```python
raw.head()
```

```python
dataset = raw

## 1.a. Age Mapper

dataset['age'] = raw['age'].map({'15-20': 0,'20-25': 1,'25-30': 2,'>35': 3})
dataset.head()
```

```python
## 1.b Sex Mapper
dataset['sex'] = raw['sex'].map({'Pria': 1,'Wanita': 0})
dataset.head()
```

```python
## 1.c Boolean Mapper ['must_use_phone', 'is_alcoholics', 'is_consuming_special_medicine']

boolean_mapper = {'Tidak': 0, 'Ya': 1}

dataset['must_use_phone'] = raw['must_use_phone'].map(boolean_mapper)
dataset.head()
```

```python
dataset['is_alcoholics'] = raw['is_alcoholics'].map(boolean_mapper)
dataset.head()
```

```python
dataset['is_consuming_special_medicine'] = raw['is_consuming_special_medicine'].map(boolean_mapper)

dataset.head()
```

```python
dataset['is_smoker'] = dataset['is_smoker'].map({'Ya, saya seorang perokok aktif': 1, 'Tidak, saya bukan seorang perokok aktif': 0})
dataset.head()
```

```python
screen_time_mapper = {
    'Kurang dari 1 jam sehari': 0,
    '1 - 2 jam sehari': 1,
    '3 - 5 jam sehari': 2,
    '6 - 8 jam sehari': 3,
    'Lebih dari 9 jam sehari': 4
}
dataset['screen_time'] = raw['screen_time'].map(screen_time_mapper)
dataset.head()
```

```python
sleep_time_mapper = {
    'Kurang dari 2 jam sehari': 0,
    '3 - 5 jam sehari': 1,
    '6 - 8 jam sehari': 2,
    'Lebih dari 9 jam sehari': 3
}
dataset['sleep_time'] = raw['sleep_time'].map(sleep_time_mapper)
dataset.head()
```

```python
dataset['is_screen_time_distract_sleep'] = raw['is_screen_time_distract_sleep'].map({'Tidak': 0, 'Mungkin': 1, 'Ya': 2})
dataset.head()
```

```python
dataset['difference_before_pandemic'] = raw['difference_before_pandemic'].map({
    'Tidak ada perubahan': 0, 
    'Ya, lebih sering menatap layar ponsel setelah pandemi': 2,
    'Ya, lebih sering menatap layar ponsel saat pandemi': 2, 
    'Ya, lebih sering menatap layar ponsel sebelum pandemi': 1
})
dataset.head()
```

```python
dataset = pd.concat([dataset, pd.get_dummies(dataset['occupation'], prefix='occupation')], axis=1)
dataset.head()
```

```python
dataset.to_csv("dataset/encoded.csv", index=False)
```
