```python
import pandas as pd
```

```python
#1 Adjusting datasets value

raw = pd.read_csv("dataset/raw.csv")
raw.shape
```

```python
raw.columns = [
    'timestamp', 'age', 'sex', 'occupation', 
    'must_use_phone', 'screen_time', 'sleep_time', 'is_smoker', 
    'is_alcoholics', 'is_consuming_special_medicine',
    'is_screen_time_distract_sleep', 'difference_before_pandemic'
]

raw.head()
```

```python
raw['difference_before_pandemic'] = raw['difference_before_pandemic'].map({
    'Tidak ada perubahan':'Tidak ada perubahan',
    'Ya, lebih sering menatap layar ponsel sebelum pandemi': 'Ya, lebih sering menatap layar ponsel sebelum pandemi',
    'Ya, lebih sering menatap layar ponsel saat pandemi': 'Ya, lebih sering menatap layar ponsel saat pandemi',
    'Ya, lebih sering menatap layar ponsel setelah pandemi': 'Ya, lebih sering menatap layar ponsel saat pandemi',
})
raw.head()
```

```python
raw = raw[raw.occupation != "Mahasiswa"]

raw = raw[raw.age != ">35"]
raw.shape
```

```python
raw.to_csv("dataset/adjusted.csv", index=False)
```
